# Análise da População Prisional de Minas Gerais (2017–2024)

## Contexto
Este projeto analisa dados públicos da população prisional do estado de Minas Gerais
entre janeiro de 2017 e junho de 2024. A análise é exploratória e descritiva,
buscando identificar padrões demográficos e sociais associados às admissões no sistema prisional.

**Fonte dos dados:** [Secretaria de Estado de Justiça e Segurança Pública de MG](https://www.seguranca.mg.gov.br/index.php/component/sppagebuilder/page/266)

## Perguntas de análise
- Como evoluiu a população prisional ao longo dos anos?
- Qual a distribuição por sexo?
- Qual o nível de escolaridade predominante?
- Há concentração em unidades ou regiões específicas?

## Observação metodológica importante
Cada linha do dataset representa uma combinação demográfica (estabelecimento + escolaridade + sexo + mês),
e a coluna `CONTAGEM_IPL` indica quantas admissões reais ocorreram naquela combinação.
**Por isso, todos os percentuais são calculados com a SOMA de `contagem_ipl`, não com contagem de linhas.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações visuais
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)
pd.set_option('display.max_columns', None)
```

```python
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_excel(
    '/content/drive/MyDrive/Notebooks_projetos/Dados Abertos_Pop_Pris_Sexo_Escolaridade_Jan_2017_Jun_2024.xlsx'
)

print(f"Dataset carregado: {df.shape[0]:,} linhas × {df.shape[1]} colunas")
df.head()
```

## 1. Entendimento inicial dos dados

Antes de qualquer análise, preciso entender:
- Quantos registros e colunas
- Tipos das variáveis
- Existência de valores nulos
- Existência de duplicatas


```python
print("DIMENSÃO DA BASE")
print(f"Linhas: {df.shape[0]:,}")
print(f"Colunas: {df.shape[1]}")
print("\nTIPOS DE DADOS")
print(df.dtypes)
print("\nVALORES NULOS")
print(df.isna().sum())
print("\nDUPLICATAS EXATAS")
print(f"Registros duplicados: {df.duplicated().sum():,}")
```

## 2. Limpeza e preparação

Etapas:
1. Remover duplicatas exatas
2. Padronizar nomes de colunas (lowercase + underscore)
3. Criar coluna de data completa (ano + mês)
4. Reagrupar categorias de escolaridade

```python
## 1Removendo duplicatas

linhas_antes = len(df)
df = df.drop_duplicates()
linhas_depois = len(df)

print(f"Registros antes:   {linhas_antes:,}")
print(f"Registros depois:  {linhas_depois:,}")
print(f"Duplicatas removidas: {linhas_antes - linhas_depois:,} "
      f"({(linhas_antes - linhas_depois) / linhas_antes * 100:.1f}% do total)")
```

```python
## 2Padronizando nomes das colunas
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(df.columns.tolist())
```

```python
##3Criar coluna Data completa
df['data_completa'] = pd.to_datetime(
    df['ano'].astype(str) + '-' + df['mes_numerico'].astype(str) + '-01',
    errors='coerce'
)

df['ano_mes'] = df['data_completa'].dt.to_period('M')

df[['ano', 'mes_numerico', 'data_completa', 'ano_mes']].head()
```

```python
## 4Reagrupar categorias de escolaridade:

# IMPORTANTE: 1.Grau = Fundamental, 2.Grau = Ensino Médio
# São níveis educacionais distintos e NÃO podem ser misturados.

df['escolaridade_na_admissao'] = (
    df['escolaridade_na_admissao']
    .str.strip()
    .str.title()
)

df['escolaridade_na_admissao'] = df['escolaridade_na_admissao'].replace({
    '1.Grau Incompleto':   'Fundamental Incompleto',
    '1.Grau Completo':     'Fundamental Completo',
    '2.Grau Incompleto':   'Médio Incompleto',
    '2.Grau Completo':     'Médio Completo',
    'Pos Doutorado':       'Pós-Doutorado',
})

# Ordem hierárquica para visualizações
ordem_escolaridade = [
    'Analfabeto',
    'Semi-Alfabetizado',
    'Fundamental Incompleto',
    'Fundamental Completo',
    'Médio Incompleto',
    'Médio Completo',
    'Superior Incompleto',
    'Superior Completo',
    'Pós-Graduado',
    'Mestrado',
    'Doutorado',
    'Pós-Doutorado',
    'Sem Informação'
]

df['escolaridade_na_admissao'].value_counts()
```

## 3. Análise de outliers — decisão metodológica

A coluna `contagem_ipl` apresenta valores extremos (algumas unidades têm volumes muito maiores que outras).
Esses outliers, no entanto, **refletem diferenças institucionais legítimas** — unidades grandes existem
e devem ser representadas. Por isso, opto por **manter os outliers** e justificar nas análises.

```python
fig, ax = plt.subplots(figsize=(10, 4))
sns.boxplot(x=df['contagem_ipl'], ax=ax, color='steelblue')
ax.set_title('Distribuição de admissões por linha (contagem_ipl)')
ax.set_xlabel('Admissões por linha')
plt.tight_layout()
plt.show()

print(f"Máximo de admissões em uma linha: {df['contagem_ipl'].max():,}")
print(f"Média:  {df['contagem_ipl'].mean():.1f}")
print(f"Mediana: {df['contagem_ipl'].median():.1f}")
```

## 4. Evolução da população prisional ao longo do tempo

Analiso o total de admissões por ano. **2024 é excluído da análise de variação**
porque o dataset cobre apenas até junho de 2024 — qualquer comparação anual seria distorcida.

```python
# Excluir 2024 (ano parcial)
df_anos_completos = df[df['ano'] < 2024]

admissoes_ano = df_anos_completos.groupby('ano')['contagem_ipl'].sum()
crescimento_pct = admissoes_ano.pct_change() * 100

crescimento_df = pd.DataFrame({
    'Admissões': admissoes_ano,
    'Variação_%': crescimento_pct.round(2)
})

crescimento_df
```

```python
#Gráfico de variação anual:
plt.figure(figsize=(10, 5))

valores = crescimento_df['Variação_%'].fillna(0).astype(float)
cores = ['#2ecc71' if x > 0 else '#e74c3c' if x < 0 else '#95a5a6' for x in valores]

bars = plt.bar(crescimento_df.index, valores, color=cores, edgecolor='black')

# Anotar valores em cima de cada barra
for bar, valor in zip(bars, valores):
    altura = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        altura + (0.3 if altura >= 0 else -0.8),
        f'{valor:.1f}%',
        ha='center',
        fontsize=10,
        fontweight='bold'
    )

plt.axhline(0, color='black', linewidth=1)
plt.title('Variação Percentual Anual das Admissões (2018–2023)', fontsize=13, fontweight='bold')
plt.ylabel('Crescimento (%)')
plt.xlabel('Ano')
plt.tight_layout()
plt.show()
```

## 5. Distribuição da população prisional por sexo

⚠️ Calculado com a **soma** de `contagem_ipl`, não com contagem de linhas.
A diferença é gigante: contar linhas dá ~76% masculino. Somar admissões reais dá ~96%.


```python
#Distribuição agregada por sexo

sexo_total = df.groupby('sexo_na_admissao')['contagem_ipl'].sum()
sexo_pct = (sexo_total / sexo_total.sum() * 100).round(2)

print("Distribuição por sexo (em % do total real de admissões):")
print(sexo_pct)
```

```python
#Gráfico de distribuição por sexo

plt.figure(figsize=(8, 5))
cores = ['#3498db', '#e91e63']
bars = sexo_pct.sort_values().plot(kind='barh', color=cores, edgecolor='black')

for i, v in enumerate(sexo_pct.sort_values()):
    plt.text(v + 1, i, f'{v}%', va='center', fontweight='bold')

plt.title('Distribuição da População Prisional por Sexo (2017–2024)', fontsize=13, fontweight='bold')
plt.xlabel('% das admissões totais')
plt.ylabel('')
plt.xlim(0, 105)
plt.tight_layout()
plt.show()
```

```python
#Evolução por sexo ao longo dos anos:

sexo_por_ano = df.pivot_table(
    index='ano',
    columns='sexo_na_admissao',
    values='contagem_ipl',
    aggfunc='sum'
)

# Converter para percentual de cada ano
sexo_por_ano_pct = sexo_por_ano.div(sexo_por_ano.sum(axis=1), axis=0) * 100
print("Distribuição por sexo, por ano (em %):")
sexo_por_ano_pct.round(2)
```

## 6. Escolaridade da população prisional

Aqui é onde a correção metodológica mais importa: agora separamos corretamente
**Ensino Fundamental** de **Ensino Médio**, que estavam fundidos antes.

```python
#Distribuição de escolaridade (soma de admissões):

escolaridade_total = (
    df.groupby('escolaridade_na_admissao')['contagem_ipl']
    .sum()
    .sort_values(ascending=False)
)

escolaridade_pct = (escolaridade_total / escolaridade_total.sum() * 100).round(2)

escolaridade_resumo = pd.DataFrame({
    'Admissões': escolaridade_total,
    '% do total': escolaridade_pct
})

escolaridade_resumo
```

```python
# Ordenar pela hierarquia educacional (não pela frequência)
ordem_presente = [c for c in ordem_escolaridade if c in escolaridade_total.index]
escolaridade_ordenada = escolaridade_total.reindex(ordem_presente)

plt.figure(figsize=(11, 6))
bars = escolaridade_ordenada.plot(kind='barh', color='#9b59b6', edgecolor='black')

# Adicionar percentuais
total = escolaridade_total.sum()
for i, (categoria, valor) in enumerate(escolaridade_ordenada.items()):
    pct = valor / total * 100
    plt.text(valor + total * 0.005, i,
             f'{valor:,} ({pct:.1f}%)', va='center', fontsize=9)

plt.title('Distribuição da Escolaridade na Admissão', fontsize=13, fontweight='bold')
plt.xlabel('Admissões (soma de contagem_ipl)')
plt.ylabel('')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
#Concentração nas faixas mais baixas:
baixa_escolaridade = [
    'Analfabeto',
    'Semi-Alfabetizado',
    'Fundamental Incompleto',
    'Fundamental Completo'
]

soma_baixa = df[df['escolaridade_na_admissao'].isin(baixa_escolaridade)]['contagem_ipl'].sum()
total = df['contagem_ipl'].sum()
pct_baixa = soma_baixa / total * 100

print(f"Admissões com escolaridade ≤ Ensino Fundamental: {soma_baixa:,}")
print(f"Percentual sobre o total: {pct_baixa:.2f}%")
```

## 7. Tipo de instituição (DEPEN vs APAC)

DEPEN é o modelo tradicional. APAC (Associação de Proteção e Assistência aos Condenados)
é um modelo alternativo, com foco em ressocialização.

```python
#Distribuição por instituição:

inst_total = df.groupby('instituicao')['contagem_ipl'].sum()
inst_pct = (inst_total / inst_total.sum() * 100).round(2)

print(pd.DataFrame({'Admissões': inst_total, '% do total': inst_pct}))
```

```python
#Cruzamento instituição × escolaridade:

cruzada = df.pivot_table(
    index='instituicao',
    columns='escolaridade_na_admissao',
    values='contagem_ipl',
    aggfunc='sum',
    fill_value=0
)

# Reordenar pela hierarquia
cruzada = cruzada.reindex(columns=[c for c in ordem_escolaridade if c in cruzada.columns])

cruzada
```

```python
# Gráfico cruzado:
cruzada.T.plot(
    kind='bar',
    stacked=True,
    figsize=(13, 6),
    colormap='tab20',
    edgecolor='black'
)
plt.title('Escolaridade na Admissão por Tipo de Instituição', fontsize=13, fontweight='bold')
plt.xlabel('Escolaridade')
plt.ylabel('Admissões')
plt.legend(title='Instituição', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 8. Concentração: ranking das unidades prisionais

Quais unidades concentram mais admissões? Uso o princípio de Pareto:
quantas unidades respondem por 50% / 80% das admissões.

```python
#Top 10 unidades:

ranking_unidades = (
    df.groupby('nome_estab_prisional')['contagem_ipl']
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
ranking_unidades.columns = ['Unidade', 'Admissões']
ranking_unidades['% do total'] = (
    ranking_unidades['Admissões'] / ranking_unidades['Admissões'].sum() * 100
).round(2)
ranking_unidades['% acumulado'] = ranking_unidades['% do total'].cumsum().round(2)

ranking_unidades.head(10)
```

```python
#Pareto: quantas unidades = 50% / 80%:

total_unidades = len(ranking_unidades)
unidades_50 = (ranking_unidades['% acumulado'] <= 50).sum()
unidades_80 = (ranking_unidades['% acumulado'] <= 80).sum()

print(f"Total de unidades distintas: {total_unidades}")
print(f"Unidades que concentram 50% das admissões: {unidades_50} "
      f"({unidades_50/total_unidades*100:.1f}% das unidades)")
print(f"Unidades que concentram 80% das admissões: {unidades_80} "
      f"({unidades_80/total_unidades*100:.1f}% das unidades)")
```

## 9. Análise regional — Regiões Integradas de Segurança Pública (RISPs)

```python
#Distribuição por RISP:

risp_total = (
    df.groupby('risp')['contagem_ipl']
    .sum()
    .sort_values(ascending=False)
)
risp_pct = (risp_total / risp_total.sum() * 100).round(2)

risp_df = pd.DataFrame({'Admissões': risp_total, '% do total': risp_pct})
risp_df
```

## 10. Indicadores consolidados

Síntese dos principais números do projeto. Todos calculados com a soma de `contagem_ipl`.

```python
#Indicadores finais:
total_admissoes = df['contagem_ipl'].sum()
total_registros = len(df)

pct_masculino = df.groupby('sexo_na_admissao')['contagem_ipl'].sum()['Masculino'] / total_admissoes * 100
pct_feminino  = df.groupby('sexo_na_admissao')['contagem_ipl'].sum()['Feminino']  / total_admissoes * 100

escolaridade_top = (
    df.groupby('escolaridade_na_admissao')['contagem_ipl']
    .sum()
    .idxmax()
)

unidade_top = (
    df.groupby('nome_estab_prisional')['contagem_ipl']
    .sum()
    .idxmax()
)

indicadores = {
    "Registros após limpeza": f"{total_registros:,}",
    "Total real de admissões (2017–jun/2024)": f"{int(total_admissoes):,}",
    "% Masculino": f"{pct_masculino:.2f}%",
    "% Feminino": f"{pct_feminino:.2f}%",
    "Escolaridade mais frequente": escolaridade_top,
    "Unidade com mais admissões": unidade_top,
    "Anos cobertos pela análise de variação": "2017–2023 (2024 excluído por ser parcial)",
}

for k, v in indicadores.items():
    print(f"{k:50s} → {v}")
```

## Conclusão

A análise da população prisional admitida em Minas Gerais entre 2017 e junho de 2024 foi conduzida sobre uma base composta por **284.627 registros** (após remoção de 44.146 duplicatas), representando aproximadamente **5,7 milhões de admissões reais** quando somada a coluna `contagem_ipl`.

### Evolução temporal
Considerando apenas anos completos (2017–2023), observa-se:
- **Crescimento** de +8,15% em 2018 e +1,45% em 2019
- **Queda expressiva** de –14,06% em 2020, coincidindo com o período inicial da pandemia de COVID-19, possivelmente refletindo políticas excepcionais de redução de aglomeração carcerária
- Oscilações pequenas entre 2021 e 2023 (–1,25%, +2,34%, –0,26%)

### Distribuição por sexo
A população prisional é fortemente predominante masculina:
- **Masculino: ~95,8%**
- **Feminino: ~4,2%**

Esse padrão se mantém estruturalmente estável ao longo do período.

### Escolaridade
A maior parte das admissões concentra-se nos níveis mais baixos de escolaridade formal. Categorias até **Ensino Fundamental Completo** somam mais da metade das admissões, evidenciando associação estrutural entre baixa escolaridade e ingresso no sistema prisional. **Importante:** essa associação é descritiva — não permite inferência causal.

### Instituições
DEPEN concentra a grande maioria das admissões, com APAC respondendo por uma fatia menor. A concentração de admissões em poucas unidades segue um padrão semelhante ao de Pareto: poucas unidades respondem pela maior parte do volume.

### Síntese
- Predominância masculina expressiva e estável
- Forte concentração em níveis educacionais mais baixos
- Queda anômala em 2020, possivelmente associada à pandemia
- Centralização das admissões em unidades específicas

### Limitações
- Ausência de variáveis socioeconômicas complementares (renda, raça/cor, ocupação)
- Inexistência de dados sobre reincidência ou tempo de permanência
- Análise exclusivamente exploratória e descritiva
- Não aplicação de modelos inferenciais ou preditivos
- 2024 com dados parciais (até junho), excluído da análise de variação anual
